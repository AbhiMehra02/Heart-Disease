### Matplotlib example workflow

`````python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# 2. setup plot
fig, ax =plt.subplots(figsize=(5,5)) #(width,height)

# 3. plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
        xlabel="x-axis",
        ylabel="y-axis")

# 5. Save & show( to save the whole figure)
fig.savefig("images/Save/sample.png")
`````

### Two Options for Subplots 

`````python
#Subplot Option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#plot to each different axis

ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
`````

`````python
# Subplot Option 2
fig,ax= plt.subplots(nrows=2,
                    ncols=2,
                    figsize=(10,5))

#plot to each differnt index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
`````

### Which one should use? (pyplot vs matplotlib OO method ?)
````
* When plotting something quickly, Okay to use the pyplot method 
* When plotting something more advanced, use the OO method

`````python
heart_disease.head()
`````

`````python
over_50=heart_disease[heart_disease["age"]>50]
over_50.size,len(over_50),over_50.shape,208*14
`````

`````python
over_50.head()
`````

`````python
#Pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             figsize=(10,6));
`````

`````python
# OO method mixed with Pyplot method
fix, ax =plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);

#ax.set_xlim([50,78]);
#ax.set_ylim([150,600])
`````

`````python
over_50.target.values
`````

`````python
## OO method from Scratch
fig, ax= plt.subplots(figsize=(10,6)) 

# Plot the data
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])

# Customiz the plot
ax.set(title="Heart disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# Add a legend 
ax.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');

# Add a Vertical line
#ax.axvline(over_50["age"].mean(),linestyle='--');
                          
`````

`````python
over_50.head()
`````

`````python
## 0. Subplots of Chol, age, and thalcah
fig,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex=True ) # to share xlabel in both plots
# 1.0 Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# 1.1 Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
        ylabel="Cholestrol")

# 1.2 Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

# 1.3 Add a horizontal line to ax0
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--")

# 2.0 Add data to ax1
scatter1=ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])
# 2.1 Customize  ax1
ax1.set(title="Heart Disease and Max Heart Rate",
             xlabel="Age",
             ylabel="Thalach")

# 2.2 Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# 2.3 Add a horizontal lline to ax1
ax1.axhline(y=heart_disease["thalach"].mean(),
            linestyle="--")

# 3. Add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
`````
